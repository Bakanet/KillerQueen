```python
def automatize(expression, word, alphabet='abc', epsilon=False):
    if epsilon:
        ctx = vcsn.context('lan_char({0}), b'.format(alphabet))
    else:
        ctx = vcsn.context('lal_char({0}), b'.format(alphabet))
        
    auto = ctx.expression(expression).thompson().proper().determinize().minimize()
    auto
```

```python
def exp(e):
    return e if isinstance(e, vcsn.expression) else ctx.expression(e)

def t1(re):
    re = exp(re)
    all = exp('[^]*')
    return all + re + all

t1('a')
```
